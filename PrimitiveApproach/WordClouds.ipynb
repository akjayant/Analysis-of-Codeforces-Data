```python
statistically_diff_tokens=[]
import scipy.stats as stats
for column in all_tokens:
    China_samples=country_token[country_token['Participant_country']=='China'][column]
    India_samples=country_token[country_token['Participant_country']=='Ð˜Ð½Ð´Ð¸Ñ'][column]
    f,p=stats.f_oneway(China_samples,India_samples)
    if p<0.05:
        statistically_diff_tokens.append((p,column,India_samples.mean(),China_samples.mean(),India_samples.var(),China_samples.var()))
```
